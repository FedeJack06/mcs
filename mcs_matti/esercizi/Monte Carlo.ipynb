```c++
TRandom3 rnd;
rnd.SetSeed(47383657);

double n=1e07;  //numero di generazioni
double a=40; //numero di facce del dado
double d1, d2,d11,d22,d;
TH1D h3("h3","",(2*a)+1,0,(2*a)+1);

for(int i=0;i<n;i++){
    d1 = rnd.Rndm()*a;
    d2 = rnd.Rndm()*a;
    
    for(int k=1;k<a;k++){
        if(k<d1 && d1<k+0.5){
            d11=k;
        }else if(k+0.5<=d1 && d1<k+1){
            d11=k+1;
        }
    }
    for(int j=0;j<a;j++){
        if(j<d2 && d2<j+0.5){
            d22=j;
        }else if(j+0.5<=d2 && d2<j+1){
            d22=j+1;
        }
    }
    d=d11+d22;
    h3.Fill(d); 
}
TCanvas c1;  c1.Draw(); h3.Draw();
```

```c++
//poisson
TRandom rnd;
rnd.SetSeed(3398427009);

double n=1e07; //numero di generazioni

TH1D h("h","",51,0.,0.);
double y, y1,y2;
int i=0;
while(i<n){
    y1=rnd.Poisson(24);
    y2=rnd.Poisson(24);
    
    y=y1+y2;
    h.Fill(y);
    i=i+1;
}
TCanvas c2; c2.Draw(); h.Draw();
```
